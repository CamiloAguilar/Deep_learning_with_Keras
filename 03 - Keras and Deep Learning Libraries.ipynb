## 2. Regression models

### Dataset de ejemplo

Digamos que nos gustaría usar `Keras` para construir rápidamente una red neuronal profunda para modelar este conjunto de datos, y así podemos automáticamente determinar la resistencia a la compresión de una muestra de concreto dada, basado en sus ingredientes.

![dataset](img/20-ejemploDF.jpg)

La red neuronal que queremos crear toma las ocho acaracterísticas como entrada, y alimenta una capa oculta de _cinco nodos_ que esta conectada a otra capa oculta de _cinco nodos_, que luego se conecta a la capa de salida con un _nodo_ que se supone sale con la resistencia a la compresión de una muestra cualquiera dada de hormigón. 

Téngase en cuenta que usualmente la solución consideraría un número mucho mayor de neuronas en cada capa oculta, al rededor de 50 o incluso 100, pero estamos usando una pequeña red para simplicidad. Nótese que todos los nodos en una capa están conectados a todos los demás nodos en la siguiente capa, lo que es conocido como una `red neuronal densa`.

![dataset](img/21-red_densa.jpg)

Antes de comenzar a usar la librería `keras`, vamos a preparar nuestrios datos y tenerlos en el formato correcto. En principio debemos dividir el `data frame` en dos, uno que tiene las columnas predictoras y otro que tiene la columna objetivo.

![dataset](img/22-particion.jpg)



```python
import pandas as pd
from sklearn.model_selection import train_test_split

concrete_data = pd.read_csv("./data/concrete_data.csv", sep=",")

old_names = concrete_data.columns
new_names = ['cement', 'Blast_Furnace_Slag', 'Fly_Ash', 'Water', 'Superplasticizer', 'Coarse_Aggregate',
            'Fine_Aggregate', 'Age', 'Strength']
concrete_data.rename(columns=dict(zip(old_names, new_names)), inplace=True)

# Dividimos el dataset en train y test
predictors = concrete_data[concrete_data.columns[:8]]
target = concrete_data[concrete_data.columns[-1]]

X_train, X_test, y_train, y_test = train_test_split(predictors, target, test_size=0.33, random_state=42)
X_train.head()
```

Veamos la magia de `keras`, usando sólo unas pocas líneas de código:

```python
import keras
from keras.models import Sequential # Puesto que nuestra red consiste de una pila lineal de capas.
from keras.layers import Dense # Para construir las capas

## Para crear el modelo usamos el constructor secuencial
model = Sequential()

n_cols = concrete_data.shape[1] - 1

## Construir las capas es bastante sencillo también, por eso importamos 'Dense'. Utilizamos add para
## agregar cada capa densa
# input_shape: Número de columnas o predictores en nuestro data frame
model.add(Dense(5, activation = "relu", input_shape=(n_cols,)))
model.add(Dense(5, activation = 'relu'))
# en la capa de salida no queremos función de activación. Además creamos un sólo nodo
model.add(Dense(1))

# Para entrenar debemos definir un optimizador y la métrica del error.
# Adam no requiere taza de aprendizaje para optimizar, lo cual es su mas grande ventaja
model.compile(optimizer='adam', loss='mean_squared_error')

## Ajustamos el modelo usando 100 iteraciones (epochs)
model.fit(X_train, y_train, epochs=500)
```

Estando completo el modelo, podemos empezar a hacer predicciones, utilizando el método `.predict`:

```python
import numpy as np
d = {'predicciones': [float(x) for x in model.predict(X_test)], 'reales': np.array(y_test)}
predictions = pd.DataFrame(data = d)
predictions.head()
```

## 3. Classification models

### Dataset de ejemplo

La base de datos de evaluación de automóviles se derivó de un modelo de decisión jerárquico simple desarrollado originalmente para la demostración de DEX, M. Bohanec, V. Rajkovic: sistema experto para la toma de decisiones. Sistemica 1 (1), pp. 145-157, 1990.). El modelo evalúa los automóviles de acuerdo con la siguiente estructura conceptual:

Aceptabilidad de coches de coches. PRECIO PRECIO GENERAL. . precio de compra de compra. . Mantenimiento del precio del mantenimiento. Características técnicas del TECH. . Comodidad confort. . . Puertas numero de puertas. . . Capacidad de personas en términos de personas para llevar. . . lug_boot el tamaño de arranque de equipaje. . Seguridad estimada de seguridad del coche.

Los atributos de entrada se imprimen en minúsculas. Además del concepto objetivo (CAR), el modelo incluye tres conceptos intermedios: PRECIO, TECHO, CONFORT. Cada concepto está en el modelo original relacionado con sus descendientes de nivel inferior mediante un conjunto de ejemplos (para estos conjuntos de conjuntos, consulte [Enlace web](https://www.kaggle.com/elikplim/car-evaluation-data-set)).

La base de datos de evaluación de automóviles contiene ejemplos con la información estructural eliminada, es decir, relaciona directamente a CAR con los seis atributos de entrada: compra, mantenimiento, puertas, personas, lug_boot, seguridad.

Debido a la estructura de concepto subyacente conocida, esta base de datos puede ser particularmente útil para probar la inducción constructiva y los métodos de descubrimiento de estructuras.


```python
## Lectura de los datos
data = pd.read_csv("./data/car_evaluation.csv", sep=",", header=None)
data.columns = ['buying_price', 'maint_price', 'num_doors', 'persons', 'lug_boot', 'safety', 'decision']
data.head()
```

```python
import pandas_profiling
pandas_profiling.ProfileReport(data)
```

```python
## Data wrangling

mapping_1 = {'vhigh': 'high', 'low': 'low', 'med': 'med', 'high': 'high'}
data.buying_price = data.buying_price.map(mapping_1)
data.maint_price = data.maint_price.map(mapping_1)

data.persons = ['2' if x == '2' else 'more' for x in data.persons]

mapping_2 = {'unacc': 0, 'acc': 1, 'good': 2, 'vgood': 3}
data.decision = data.decision.map(mapping_2)

car_data = pd.concat([pd.get_dummies(data[['buying_price', 'maint_price', 'persons']]), data[['decision']] ], axis=1)
car_data.head()
```

A continuación dividimos nuestro conjunto de datos en predictores y objetivo. Sin embargo con `keras`, por problemas de clasificación no podemos usar la columna objetivo como tipo `pandas Series`. La transformamos en una matriz con valores binarios, similar al `one hot encoding`, usando la función `to_categorical`

```python
from keras.utils import to_categorical

predictors = car_data[car_data.columns[:8]]
target = to_categorical(car_data['decision'])

X_train, X_test, y_train, y_test = train_test_split(predictors, target, test_size=0.33, random_state=42)
X_train.head()
```

Aplicamos el modelo de clasificación en `Keras`, teniendo en cuenta que ahora consideramos cuatro categorías en vez de una.

![dataset](img/23-classification.jpg)

```python
import keras
from keras.models import Sequential
from keras.layers import Dense

model = Sequential()

n_cols = predictors.shape[1]

## Definimos la arquitectura de la red
model.add(Dense(32, activation='relu', input_shape=(n_cols,)))
model.add(Dense(20, activation='relu'))
model.add(Dense(32, activation='relu'))
model.add(Dense(4, activation='softmax'))

## Compilamos el modelo
model.compile(optimizer='adam', loss='categorical_crossentropy',
             metrics=['accuracy'])

## Ajustamos el modelo usando 500 iteraciones (epochs)
model.fit(X_train, y_train, epochs=500)
```

Revisamos las predicciones sobre el conjunto de `test`

```python
predicciones = model.predict(X_test)

pred = pd.Series([np.argmax(x) for x in predicciones], name='Pred')
prob = pd.Series([np.max(x) for x in predicciones], name='Prob')
real = pd.Series([np.argmax(x) for x in y_test], name='real')

resultados = pd.concat([pred, prob, real], axis=1)
resultados.head()
```

```python
from sklearn.metrics import confusion_matrix

print(real.value_counts())
confusion_matrix(pred, real)
```
